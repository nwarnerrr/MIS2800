```python
import sqlite3
print("sqlite3 imported")
```

```python
conn = sqlite3.connect('AdCampaigns2026.db')
print("database created")
```

```python
%sql sqlite:///AdCampaigns2026.db
```

```python
#Block 1
%%sql
CREATE TABLE Marketing_Brands (
  brand_id INTEGER PRIMARY KEY,
  brand_name TEXT,
  brand_type TEXT,
  brand_budget INTEGER
) STRICT;
```

```python
#Block 2
%%sql
INSERT INTO Marketing_Brands (brand_id, brand_name, brand_type, brand_budget)
VALUES
  (1, 'Doritos', 'Snacks', 7000000),
  (2, 'AWS', 'Technology', 15000000),
  (3, 'Pepsi', 'Beverages', 2000000),
  (4, 'He_Gets_Us', 'Religious', 1000000),
  (5, 'Dr_Pepper', 'Beverages', 5000000),
  (6, 'IBM', 'Technology', 100000),
  (7, 'Google', 'Technology', 8000000);
```

```python
#Block 3

%%sql
SELECT * FROM Marketing_Brands;
```

```python
#Block 4
%%sql
SELECT brand_name, brand_type
FROM Marketing_Brands
WHERE brand_budget > 700000;
```

```python
#Block 5
%%sql
SELECT brand_name
FROM Marketing_Brands
WHERE brand_type = 'Religious';
```

```python
#Block 6
%%sql
SELECT brand_name
FROM Marketing_Brands
WHERE brand_budget < 1000000;
```

```python
#Block 7
query = """
SELECT brand_name
FROM Marketing_Brands
WHERE brand_budget > 7000000
"""
results = conn.execute(query)

for row in results:
    print(row)
```

```python
#Block 8
%%sql
SELECT brand_name
FROM Marketing_Brands
WHERE (brand_type = 'Snacks' OR brand_type = 'Beverages')
```

```python
#Block 9
%%sql
SELECT *
FROM Marketing_Brands
ORDER BY brand_budget DESC;
```

```python
#Block 10
query = "SELECT * FROM Marketing_Brands ORDER BY brand_budget ASC"
results = conn.execute(query)
for row in results:
  print(row)
```
